```python
# 流动
flow = np.ones([2,11,11])
flow

pop = np.full(11,2000)

initial = np.random.rand(11)
n = 11
iterations = 60

r = 2

R0 = 2.4
alpha = np.ones(11)
lengthOfIncubationPeriod = 4
lengthOfTreatment  = 7
```

```python
Svec = pop.copy() - initial
Evec = np.zeros(n)
Ivec = initial
Avec = np.zeros(n)
Rvec = np.zeros(n)
Qvec = np.zeros(n)
res = np.zeros((iterations, 4))
history = np.zeros((iterations, 4, n))
history[0, 0, :] = Svec
history[0, 1, :] = Evec
history[0, 2, :] = Ivec
history[0, 3, :] = Rvec
N = pop.sum()
test = True

for i in range(0, iterations - 1):

    realOD = flow[i % r]
    #新的病例，潜伏，康复
    newE = Svec * Ivec * R0 * alpha
    newI = Evec / lengthOfIncubationPeriod
    newR = Ivec / lengthOfTreatment
    if i > 0:
        #人口流动的系数比
        realOD = flow[1 % r]
        row_sums = initial.sum()
        Ivec_n = Ivec / row_sums
        infected_mat = np.array([Ivec_n, ] * 11).transpose()
        infected_mat = np.log(infected_mat + 1)
        OD_infected = np.round(realOD * infected_mat)
        inflow_infected = OD_infected.sum(axis=0) * 0.02

        #         peopleInfectedNormal = Svec * inflow_infected * R0 / (pop + realOD.sum(axis=0)) * a
        #         peopleInfectedNormal = np.where( peopleInfectedNormal>Svec, Svec,  peopleInfectedNormal)
        #         peopleInfectedNormal = np.round(peopleInfectedNormal)
        #         if i >5:
        #             peopleInfectedNormal = Svec * 0.3 * eff * alpha
        eff = 0.543
        peopleInfectedNormal = Svec * 0.3 * eff * alpha
        peopleconvert = Evec / lengthOfIncubationPeriod
        peopleconvert = np.round(peopleconvert)

        peoplerecover = Ivec / lengthOfTreatment

        Svec = Svec - peopleInfectedNormal
        Svec = np.round(Svec)

        Evec = Evec + peopleInfectedNormal - peopleconvert
        Evec = np.round(Evec)

        Ivec = Ivec + peopleconvert - peoplerecover
        Ivec = np.round(Ivec)
        Rvec = Rvec + peoplerecover
        Rvec = np.round(Rvec)

        #         if (i>45):
        #             test = False
        #         if (i>50):
        #             Ivec = Ivec * 0.2

        res[i + 1, :] = [Svec.sum(), Evec.sum(), Ivec.sum(), Rvec.sum()]
        eachIter = np.zeros(iterations + 1)
        eachIter[i + 1] = newI.sum()
        history[i + 1, 0, :] = Svec
        history[i + 1, 1, :] = Evec
        history[i + 1, 2, :] = Ivec
        history[i + 1, 3, :] = Rvec
```
